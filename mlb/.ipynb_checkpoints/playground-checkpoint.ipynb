```python
from datetime import date

today = date.today().strftime('%m/%d/%Y')

print(today)
```

```python
import statsapi
from pprint import pprint as pp

sched = statsapi.schedule(date=f"{today}")

pp(sched)
```

```python
leaders = statsapi.league_leaders('strikes',statGroup='pitching',limit=10)
print(leaders)
```
